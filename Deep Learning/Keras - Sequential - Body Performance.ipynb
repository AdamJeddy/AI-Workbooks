Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import OrdinalEncoder
import tensorflow as tf
from tensorflow.keras import layers
```

### Import Data

```python
df = pd.read_csv('bodyPerformance.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
class_names = ['1', '2', '3', '4']
```

```python
enc = OrdinalEncoder()
temp = enc.fit_transform(df)
temp = pd.DataFrame(temp)
```

```python
features = temp.copy()
labels = features[features.columns[:-1]]
```

```python
features.head()
```

```python
labels.head()
```

```python
features_arr = np.array(features)
features_arr
```

```python
model = tf.keras.Sequential([
  layers.Dense(64),
  layers.Dense(11)
])

model.compile(loss = tf.keras.losses.MeanSquaredError(), 
              optimizer = tf.optimizers.Adam(), 
              metrics=['accuracy'])
```

```python
features_arr = np.asarray(features_arr).astype(np.float32)
```

```python
model.fit(features_arr, labels, epochs=15)
```

```python
normalize = layers.Normalization()
```

```python
normalize.adapt(features_arr)
```

```python
norm_abalone_model = tf.keras.Sequential([
  normalize,
  layers.Dense(64),
  layers.Dense(11)
])

norm_abalone_model.compile(loss = tf.losses.MeanSquaredError(),
                           optimizer = tf.optimizers.Adam(),
                           metrics=['accuracy'])

norm_abalone_model.fit(features_arr, labels, epochs=15)
```
